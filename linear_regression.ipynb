```python
import numpy as np
import matplotlib.pyplot as plt

xi=np.array([1,2,3,4,5])
yi=np.array([1.2,1.8,2.6,3.2,3.8])
```

```python
x_mean=np.mean(xi)
y_mean=np.mean(yi)
```

```python
numerator = np.mean(xi*yi)-x_mean *y_mean
denominator =np.mean(xi**2)-x_mean**2
a1=numerator / denominator
```

```python
a0 = y_mean - a1 * x_mean
```

```python
week_7 =7
week_12 =12
sales_7 =a0+a1*week_7
sales_12 =a0+a1*week_12
```

```python
print("predicted sales for the 7th week : {:.2f} thousand units".format(sales_7))
print("predicted sales for the 12th week : {:.2f} thousand units".format(sales_12))
```

```python
plt.scatter(xi,yi,label="Data points")
plt.plot(xi,a0+a1*xi,color='red',label='regression line')
plt.xlabel('week')
plt.ylabel('sales (in thousand)')
plt.title('simple linear regression')
plt.legend()
plt.show()
```
